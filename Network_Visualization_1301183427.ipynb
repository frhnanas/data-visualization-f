```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(18, 10))
G = nx.Graph()
G.add_edge("BANDUNG", "TASIKMALAYA")
G.add_edge("BANDUNG", "SUBANG")
G.add_edge("BANDUNG", "CIMAHI")
G.add_edge("BANDUNG", "SOREANG")
G.add_edge("BANDUNG", "GARUT")
G.add_edge("BANDUNG", "SUMEDANG")
G.add_edge("BEKASI", "KARAWANG")
G.add_edge("CIAMIS", "TASIKMALAYA")
G.add_edge("CIBINONG", "BOGOR")
G.add_edge("CIMAHI", "CIANJUR")
G.add_edge("CIMAHI", "PURWAKARTA")
G.add_edge("CIREBON", "INDRAMAYU")
G.add_edge("CIREBON", "KARAWANG")
G.add_edge("CIREBON", "KUNINGAN")
G.add_edge("CIANJUR", "BOGOR")
G.add_edge("CIANJUR", "SUKABUMI")
G.add_edge("CIANJUR", "SOREANG")
G.add_edge("GARUT", "TASIKMALAYA")
G.add_edge("KARAWANG", "PURWAKARTA")
G.add_edge("KARAWANG", "INDRAMAYU")
G.add_edge("KUNINGAN", "SUMEDANG")
G.add_edge("SUMEDANG", "MAJALENGKA")
G.add_edge("SUMEDANG", "PURWAKARTA")
G.add_edge("SUKABUMI", "DEPOK")
nx.draw_networkx(G, with_labels=True)
plt.show()
```

```python
degree = sorted(nx.degree_centrality(G), key=nx.degree_centrality(G).get, reverse=True)
betweenness = sorted(nx.betweenness_centrality(G), key=nx.betweenness_centrality(G).get, reverse=True)
print("Kota terpenting berdasarkan degree centrality:",degree[0],
      "dan betweenness centrality:", betweenness[0])
```
