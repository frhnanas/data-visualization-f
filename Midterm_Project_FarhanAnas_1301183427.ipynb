### Soal 3 (30 poin)
Lakukanlah _trend analysis_ nilai _daily percentage change_ dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan melalui visualisasi data.
_Trend analysis_ dilakukan dengan menggunakan hubungan berikut:
![output%2011.png](https://www.dl.dropboxusercontent.com/s/dxmrm74ksz5l9v2/trend_analysis_fig.png?dl=0)

#### Code

```python
# ambil data Hang_Seng.csv
df_HS = pd.read_csv('Hang_Seng.csv')
# bulatkan menjadi 2 angka dibelakang koma
df_HS = df_HS.round(2)
# hilangkan data bernilai nil
df_HS.dropna(inplace = True, axis = 0)
# ubah tipe data date
df_HS['Date'] = pd.to_datetime(df_HS['Date'])
df_HS.head()
```

```python
# ambil data Nasdaq.csv
df_ND = pd.read_csv('Nasdaq.csv')
# bulatkan menjadi 2 angka dibelakang koma
df_ND = df_ND.round(2)
# hilangkan data bernilai nil
df_ND.dropna(inplace = True, axis = 0)
# ubah tipe data date
df_ND['Date'] = pd.to_datetime(df_ND['Date'])
df_ND.head()
```

```python
# ambil data Nikkei.csv
df_NK = pd.read_csv('Nikkei.csv')
# bulatkan menjadi 2 angka dibelakang koma
df_NK = df_NK.round(2)
# hilangkan data bernilai nil
df_NK.dropna(inplace = True, axis = 0)
# ubah tipe data date
df_NK['Date'] = pd.to_datetime(df_NK['Date'])
df_NK.head()
```

```python
# ubah menjadi hanya tahun 2019 saja sesuai soal
df_HS.index = df_HS['Date']
df_HS = df_HS.loc['2019-01-01':'2019-12-31']
df_ND.index = df_ND['Date']
df_ND = df_ND.loc['2019-01-01':'2019-12-31']
df_NK.index = df_NK['Date']
df_NK = df_NK.loc['2019-01-01':'2019-12-31']
```

```python
# hitung day percent change menggunakan adjusted closing price lalu diassign ke dalam kolom Day_Perc_Change
df_HS['Day_Perc_Change'] = df_HS['Adj Close'].pct_change()*100
df_HS.dropna(inplace = True, axis = 0)
df_ND['Day_Perc_Change'] = df_ND['Adj Close'].pct_change()*100
df_ND.dropna(inplace = True, axis = 0)
df_NK['Day_Perc_Change'] = df_NK['Adj Close'].pct_change()*100
df_NK.dropna(inplace = True, axis = 0)
```

```python
df_HS.head()
```

```python
df_ND.head()
```

```python
df_NK.head()
```

```python
# fungsi trend analysis menggunakan tabel diatas
def trend(t):
  if t > -0.5 and t <= 0.5:
    return "Slight or No Change"
  elif t > 0.5 and t <= 1:
    return "Slight Positive"
  elif t > -1 and t <= -0.5:
    return "Slight Negative"
  elif t > 1 and t <= 3:
    return "Positive"
  elif t > -3 and t <= -1:
    return "Negative"
  elif t > 3 and t <= 7:
    return "Among Top Gainers"
  elif t > -7 and t <= -3:
    return "Among Top Losers"
  elif  t > 7:
    return "Bull Run"
  elif t <= -7:
    return "Bear Drop"
```

```python
# hitung trend menggunakan day percent change lalu diassign ke dalam kolom trend
df_HS['Trend'] = df_HS['Day_Perc_Change'].apply(lambda t:trend(t))
df_ND['Trend'] = df_ND['Day_Perc_Change'].apply(lambda t:trend(t))
df_NK['Trend'] = df_NK['Day_Perc_Change'].apply(lambda t:trend(t))
```

```python
df_HS.head()
```

```python
df_ND.head()
```

```python
df_NK.head()
```

```python
# visualisasi ketiga dataset dengan menggunakan pie chart mengacu pada trend
plt.figure(1)
HS_data = df_HS.groupby('Trend')
pie_label_HS = sorted([i for i in df_HS.loc[:, 'Trend'].unique()])
plt.pie(HS_data['Trend'].count(), labels = pie_label_HS, autopct = '%1.1f%%', radius = 2)
plt.title('Hang Seng', y = 1.5)

plt.figure(2)
ND_data = df_ND.groupby('Trend')
pie_label_ND = sorted([i for i in df_ND.loc[:, 'Trend'].unique()])
plt.pie(ND_data['Trend'].count(), labels = pie_label_ND, autopct = '%1.1f%%', radius = 2)
plt.title('Nasdaq', y = 1.5)

plt.figure(3)
NK_data = df_NK.groupby('Trend')
pie_label_NK = sorted([i for i in df_NK.loc[:, 'Trend'].unique()])
plt.pie(NK_data['Trend'].count(), labels = pie_label_NK, autopct = '%1.1f%%', radius = 2)
plt.title('Nikkei', y = 1.5)

plt.show()
```

#### Penjelasan 
Pada visualisasi data ketiga pie chart diatas :

Saham Hang Seng memiliki trend negatif yang paling banyak di tahun 2019 dimana ditandai dengan nilai 11.8% (lebih tinggi dibandingkan dengan saham lainnya) dan berada di tengah-tengah untuk trend positifnya yakni dengan nilai 13.1% dan saham ini juga sempat menjadi yang terbaik ditandai dengan nilai 0.4% pada Top Gainers.

Saham Nasdaq memiliki trend positif yang paling banyak di tahun 2019 dimana ditandai dengan nilai  17.1% (lebih tinggi dibandingkan dengan saham lain) dan berada di tengah-tebgah untuk trend negatifnya yakni dengan nilai 9.2% dan saham ini juga sempat menjadi yang terbaik sekaligus yang terburuk ditandai dengan nilai 0.4% pada Top Gainers dan nilai 1.6% pada Top Losers.

Saham Nikkei merupakan saham yang paling stabil ditandai dengan nilai 52.9% pada Slight or No Change akan tetapi pernah menjadi yang terburuk dengan nilai 0.4% pada Top Losers.

### Soal 4 (30 poin)
Lakukanlah analisis volatilitas dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan hasil visualisasinya.

#### Code

```python
# buat dataframe gabungan dari ketiga dataset diatas
df_mix = pd.concat([df_HS['Adj Close'], df_ND['Adj Close'], df_NK['Adj Close']], axis = 1)
df_mix.columns = ['Hang_Seng', 'Nasdaq', 'Nikkei']
# hilangkan data bernilai nil
df_mix.dropna(inplace = True, axis = 0)
df_mix.head()
```

```python
# ubah data dalam bentuk percent dan diassign ke dalam dataframe baru
df_percent = df_mix.pct_change()*100
# hilangkan data bernilai nil
df_percent.dropna(inplace = True, how = 'any', axis = 0)
df_percent.head()
```

```python
# visualisasi volatilitas dari dataframe yang sudah dibuat
volatile = df_percent[['Hang_Seng', 'Nasdaq', 'Nikkei']].rolling(7).std()*np.sqrt(7)
volatile.plot(figsize = (20, 20))
```

#### Penjelasan 
Pada visualisasi data volatilitas diatas :

Saham Hang Seng memiliki nilai volatilitas yang paling stabil sepanjang tahun 2019 ditandai dengan nilai yang tidak terlalu tinggi dan juga tidak terlalu rendah.

Saham Nasdaq memiliki nilai volatilitas yang paling tidak stabil yakni dalam artian sering mengalami kenaikan bahkan sempat menjadi yang tertinggi di quartil ke 3 tahun 2019 namun sering juga mengalami penurunan yakni pada quartil ke 2 dan 4 pada tahun 2019.

Saham Nikkei memiliki nilai volatilitas yang cenderung tinggi di quartil ke 1 tahun 2019 akan tetapi lama kelamaan mengalami penurunan dan sempat menjadi yang terendah pada bulan 11.
